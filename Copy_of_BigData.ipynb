```python
# 2. Baca file Excel
df = pd.read_excel("data_lalu_lintas.xlsx")

print("Data Awal:")
print(df)

# 3. DATA CLEANING
# a. Hapus duplikat
df = df.drop_duplicates()

# b. Hilangkan spasi berlebih
df["Sumber"] = df["Sumber"].str.strip()
df["Kondisi"] = df["Kondisi"].str.strip()

# c. Samakan format huruf
df["Kondisi"] = df["Kondisi"].str.capitalize()

# d. Perbaiki typo
df["Keterangan"] = df["Keterangan"].str.replace("tingi", "tinggi")

# e. Format jam
df["Jam"] = pd.to_datetime(df["Jam"], format="%H:%M").dt.time

# 4. PREPROCESSING

# Encoding kondisi
mapping = {"Lancar": 1, "Padat": 2, "Macet": 3}
df["Kondisi_Encoded"] = df["Kondisi"].map(mapping)

# 5. CEK HASIL
print("\nJumlah Data:", len(df))
print("\nData Setelah Cleaning:")
print(df)

# 6. SIMPAN KE EXCEL
df.to_excel("data_bersih.xlsx", index=False)

print("\nFile data_bersih.xlsx berhasil dibuat!")
```
